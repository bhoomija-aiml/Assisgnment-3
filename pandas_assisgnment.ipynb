```python
# part2
import pandas as pd
df = pd.read_csv('world_population.csv')
df
```

```python
# Q1
print("Task 1: Top 10 rows of the dataset:")
print(df.head(10))
```

```python
# Q2
print("\nNull Values in Each Column:")
print(df.isnull().sum())
df.dropna(inplace=True)
```

```python
# Q3
df["2022 Population"] = pd.to_numeric(df["2022 Population"], errors='coerce')
top_5_populous = df.nlargest(5, "2022 Population")[["Country/Territory", "2022 Population"]]
print("\nTop 5 Most Populous Countries:")
print(top_5_populous)
```

```python
# Q4
if "GDP" in df.columns:
    df["GDP per Capita"] = df["GDP"] / df["2022 Population"]
    print("\nGDP per Capita added.")
else:
    print("\n'GDP' column not found â€” skipping GDP per Capita calculation.")
```

```python
# Q5
filtered_countries = df[(df["2022 Population"] > 100_000_000) & (df["Area (kmÂ²)"] < 500_000)]
print("\nCountries with Population > 100 million and Area < 500,000 kmÂ²:")
print(filtered_countries[["Country/Territory", "2022 Population", "Area (kmÂ²)"]])
```

```python
# Q6
df_sorted_by_density = df.sort_values(by="Density (per kmÂ²)", ascending=False)
print("\nTop 10 Countries by Population Density:")
print(df_sorted_by_density[["Country/Territory", "Density (per kmÂ²)"]].head(10))
```

```python
# part3
import pandas as pd
# Q1
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
matches
deliveries
```

```python
# Q2
most_successful_team = matches['winner'].value_counts().idxmax()
most_wins = matches['winner'].value_counts().max()
print(f"Most successful team: {most_successful_team} with {most_wins} wins")
```

```python
# Q3
matches_per_season = matches['season'].value_counts().sort_index()
print("\nMatches per Season:")
print(matches_per_season)
```

```python
# Q4
top_players = matches['player_of_match'].value_counts().head(5)
print("\nTop 5 Players with Most 'Player of the Match' Awards:")
print(top_players)
```

```python
# Q5
merged_df = deliveries.merge(matches, left_on='match_id', right_on='id')
# Example analysis from merged data:
# Top 5 run scorers overall
top_scorers = merged_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Run Scorers Overall:")
print(top_scorers)
```

```python
# part4
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
# Q1
avg_math = df['math score'].mean()
avg_reading = df['reading score'].mean()
avg_writing = df['writing score'].mean()

print("Average Math Score:", round(avg_math, 2))
print("Average Reading Score:", round(avg_reading, 2))
print("Average Writing Score:", round(avg_writing, 2))
```

```python
# Q2
gender_avg_writing = df.groupby('gender')['writing score'].mean()
print("\nAverage Writing Score by Gender:")
print(gender_avg_writing)

better_writer = gender_avg_writing.idxmax()
print(f"\n{better_writer.capitalize()}s performed better in writing on average.")
```

```python
# Q3
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['percentage'] = df['total score'] / 3
```

```python
# Q4
grouped_avg = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

grouped_avg.plot(kind='bar', figsize=(12, 6))
plt.title('Average Scores by Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Subject')
plt.show()
```

```python
# Q5
top_students = df[(df['math score'] >= 90) & (df['reading score'] >= 90) & (df['writing score'] >= 90)]
print(f"\nNumber of students who scored above 90 in all subjects: {len(top_students)}")
print(top_students[['gender', 'math score', 'reading score', 'writing score', 'percentage']])
```
